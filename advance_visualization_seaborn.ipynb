```python
import seaborn as sns
```

```python
df = sns.load_dataset('flights')
```

```python
df.head()
```

```python
df = df.pivot('month', 'year', 'passengers')
```

```python
df
```

```python
sns.heatmap(df)
```

```python
sns.clustermap(df)
```

```python
sns.clustermap(df, col_cluster=False)
```

```python
sns.clustermap(df, row_cluster=False)
```

```python

sns.clustermap(df, cmap='coolwarm', linewidths=1)
```

```python
sns.clustermap(df, cmap='Blues_r', linewidths=1.4)
```

```python
sns.clustermap(df, standard_scale = 1)
```
